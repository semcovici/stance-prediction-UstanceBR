```python
df_results_macro = df_results[df_results['class'] == 'macro avg'].sort_values('f1-score')
df_results_macro.head()
```

```python
dict_cp = {
    'cl':'3. Hydrox.',
    'lu':'1. Lula',
    'co':'4. Sinovac',
    'ig':'6. Church',
    'gl':'5. Globo TV',
    'bo':'2. Bolsonaro',
}
```

```python
df_results_macro.file_name.sort_values().unique()
```

```python
for file_name in df_results_macro.file_name.sort_values().unique():
    
    print(file_name)
    df_results_macro_c = df_results_macro[df_results_macro.file_name == file_name][['corpus','precision','recall' ,'f1-score']]
    df_results_macro_c.corpus = df_results_macro_c.corpus.map(dict_cp)
    df_results_macro_c.sort_values('corpus', inplace=True)
    
    df_results_macro_c.loc[len(df_results_macro_c)] = {
        'corpus':'overall',
        'precision':df_results_macro_c.precision.mean(),
        'recall':df_results_macro_c.recall.mean(),
        'f1-score':df_results_macro_c['f1-score'].mean(),
    }
    
    df_results_macro_c = df_results_macro_c.round(2)
    display(df_results_macro_c)
    print(' & '.join(df_results_macro_c['f1-score'].astype('str').tolist()))  
    
    print('\n\n\n\n')  
```
