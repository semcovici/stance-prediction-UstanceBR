```python
import pandas as pd
from tsfresh.feature_extraction import MinimalFCParameters
from tsfresh.feature_extraction import extract_features
from tqdm import tqdm
tqdm.pandas() 
import numpy as np
```

```python
settings = MinimalFCParameters()
```

```python

path_processed_data = '../data/processed/'
path_interim_data = '../data/interim/'
model_name = 'facebook/fasttext-pt-vectors'
model_name = f'neuralmind/bert-base-portuguese-cased'
corpus = 'ig'
path_data = path_interim_data + f'train_r3_{corpus}_separated_comments_{model_name.replace('/', '_')}_part_1.parquet'

emb_dim = 300
```

```python
data = pd.read_parquet(path_data) 
```

```python
data
```

```python
data.head()
```

```python
data_emb = data.drop(
    ['Texts', 'Polarity'],
    axis =1
)
```

```python
data_emb.head()
```

```python
X_tsfresh = extract_features(
    data_emb,
    column_id='User_ID',
    default_fc_parameters= settings
)
```

```python
X_tsfresh.head()
```

```python
X_tsfresh.shape
```

```python
X_tsfresh_cplt = X_tsfresh.reset_index(drop=False).rename(columns={'index': 'User_ID'}).merge(
    data[['User_ID','Polarity']].drop_duplicates(),
    on = 'User_ID',
    how = 'left'
)
```

```python
path_output = path_processed_data + f'train_r3_{corpus}_separated_comments_{model_name.replace("/", "_")}_tsfresh.parquet'

X_tsfresh_cplt.to_parquet(path_output,index = False)
```
